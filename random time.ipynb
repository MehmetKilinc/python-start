```python
import random 

import time


rasgele_sayı = random.randint(1,40)

tahmin_hakkı = 7

print("çıkmak için q ya basın")

while True:
    
    tahmin_edilen_sayı = int(input("tahmininizi giriniz :"))
    
    
    if (tahmin_edilen_sayı < 1 or tahmin_edilen_sayı > 40):
        
        print("lütfen 1 ve 40 arasında bir sayıyı giriniz")
        
    
    elif (tahmin_edilen_sayı == rasgele_sayı):
        
        print("cevap aranıyor...")
        
        time.sleep(2)
        
        print("tebrikler kazandınız")
        
    elif (tahmin_edilen_sayı < rasgele_sayı):
        
        print("cevap aranıyor")
        
        time.sleep(1)
        
        print("daha büyük bir sayı girmelisiniz")
        
        tahmin_hakkı -= 1
        
    elif (tahmin_edilen_sayı > rasgele_sayı):
        
        print("cevap aranıyor")
        
        time.sleep(1)
        
        print("daha küçük bir sayı girmelisiniz")
        
        tahmin_hakkı -= 1
        
        
    elif (tahmin_edilen_sayı == "q" or tahmin_edilen_sayı == "Q"):
        
        print("hoşçakalın . tekrar bekleriz")
        
        break
        
    else:
        
        print("hatalı bir tuşa bastınız")
        
        break
```
