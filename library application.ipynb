# library application


```python
import os
BookList = list()

menu = """

[1] add book
[2] take book
[3] book list
[Q] exit

"""
def addBook(book : tuple , lista : list):
    lista.append(book)
    print("adding process completed...")
    print("Press enter to return to the menu")
    input()

def control(book : tuple , lista : list):
    if book in lista:
        return True
    else:
        return False

    
def deleteBook(book : tuple , lista : list):
    if control(book , lista):
        lista.remove(book)
        print("book deletion completed...")
        print("press enter to return to the menu")
    else:
        print("the book you asked is not available")
        print("press enter to return to the menu")
        
        
def toTheList(lista : list):
    for i in lista:
        print("book name : {} -------------->>>>>>>  author name : {}".format(i[0],i[1]))
    print("press enter to return to the menu")
    input()
    
while True:
    os.system("cls")
    print(menu)
    
    selection = input("choose action : ")
    
    if selection == "1":
        book_name = input("enter book name :")
        author_name = input("enter author name")
        book = (book_name , author_name)
        addBook(book , BookList)
        
    elif selection == "2":
        book_name = input("enter book name :")
        author_name = input("enter author name")
        book = (book_name , author_name)
        deleteBook(book , BookList)
        
    elif selection == "3":
        toTheList(BookList)
        
    elif selection == "q" or "Q":
        print("pleasant readings...")
    
    else:
        print("you entered wrong")
    
```
