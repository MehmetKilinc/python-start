```python
import os

dictionary = {"book" : "kitap" , "cyber securtiy" : "siber güvenlik" , "computer" : "bilgisayar" , "pen" : "kalem"}


menu = """

[1] see all words

[2] search word

[3] exit

"""
while True:
    os.system("cls")
    print(menu)
    
    selection = input("choose action")
    
    if selection == "1":
        for i in dictionary.items():
            print(i)
            
        print("press enter to exit")
        input()
        
    elif selection == "2":
        word_query = input("type the word you want to query")
        print(dictionary[word_query])
        print("press enter to exit")
        input()
        
    elif selection == "3":
        print("We hope you come again")
        input()
        
    else:
        print("you entered incorrectly")
        
        
        
        
        
```
